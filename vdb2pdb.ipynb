```python
import re
import string

vdb_file = '1ncr_full.vdb'
CHAIN_LIST = list(string.ascii_uppercase)

current_chain = None
chain_count = 0  
output_prefix = "".join(vdb_file.split('.')[:-1])
file_count = 0
filename = None

with open(vdb_file, 'r') as f:
    for line in f:              
        if re.match("^ATOM", line) or re.match('^HETATM', line):  
            chain = line[21]  # get the chain ID

            # Check if the chain has changed
            if not current_chain or chain != current_chain:
                current_chain = chain
                chain_count += 1 
                
                # open a new file for each set of 26 chains
                if (chain_count - 1) % 26 == 0:
                    file_count+=1
                    if filename:
                        output_file.close()
                    
                    filename = "{}_{}{}".format(output_prefix, file_count, ".pdb")
                    print("\n",filename)
                
                    output_file = open(filename, 'w')
                
                output_chain = CHAIN_LIST[(chain_count - 1) % 26]
                line = line[:21]+ current_chain + line[21:]
#                 print (CHAIN_LIST[(chain_count - 1) % 26], end=' ')

            output_file.write(line)
        else:
            print(line)


if filename:
    output_file.close()            
print("completed") 
```
